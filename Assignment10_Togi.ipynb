1. Tunjukan scatter plot untuk kombinasi 2 feature ini:
  - SepalLengthCm dan PetalLengthCm
  - SepalWidthCm dan PetalWidthCm
  - SepalLengthCm dan PetalWidthCm

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = pd.read_csv('iris.csv')
dataset.drop(['Id', 'Species'], axis=1, inplace=True)
```

```python
dataset.head()
```

```python
from sklearn.cluster import KMeans
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=10)
    kmeans.fit(dataset)
    wcss.append(kmeans.inertia_)

print(wcss)
```

```python
#Elbow Method
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
```

```python
kmeans = KMeans(n_clusters = 3, random_state = 10)
model = kmeans.fit(dataset)
```

```python
model.labels_
```

```python
model.cluster_centers_
```

```python
y_kmeans = kmeans.predict(dataset)
```

```python
#Scatter Plot Sepal Length dan Petal Length
plt.scatter(dataset.values[y_kmeans == 0, 0], dataset.values[y_kmeans == 0, 2], s=70, c='red', label ='Cluster 1')
plt.scatter(dataset.values[y_kmeans == 1, 0], dataset.values[y_kmeans == 1, 2], s=70, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[y_kmeans == 2, 0], dataset.values[y_kmeans == 2, 2], s=70, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 2], s=50, c='yellow', label = 'Centroids')
```

```python
#Scatter Plot Sepal Width dan Petal Width
plt.scatter(dataset.values[y_kmeans == 0, 1], dataset.values[y_kmeans == 0, 3], s=70, c='red', label ='Cluster 1')
plt.scatter(dataset.values[y_kmeans == 1, 1], dataset.values[y_kmeans == 1, 3], s=70, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[y_kmeans == 2, 1], dataset.values[y_kmeans == 2, 3], s=70, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 1], model.cluster_centers_[:, 3], s=50, c='yellow', label = 'Centroids')
```

```python
#Scatter Plot Sepal Width dan Petal Width
plt.scatter(dataset.values[y_kmeans == 0, 0], dataset.values[y_kmeans == 0, 3], s=70, c='red', label ='Cluster 1')
plt.scatter(dataset.values[y_kmeans == 1, 0], dataset.values[y_kmeans == 1, 3], s=70, c='blue', label ='Cluster 2')
plt.scatter(dataset.values[y_kmeans == 2, 0], dataset.values[y_kmeans == 2, 3], s=70, c='green', label ='Cluster 3')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 3], s=50, c='yellow', label = 'Centroids')
```

2. Untuk dataset **customer**
  - Cari nilai k yang paling optimal menggunakan Elbow method
  - Buat scatter plot untuk kombinasi 2 feature ini:
    - Age dan Annual Income
    - Annual Income dan Spending Score
    - Age dan Spending Score

```python
datacustomer = pd.read_csv('customer.csv')
datacustomer.head(10)
```

```python
datacustomer.drop(['CustomerID'], axis = 1, inplace=True)
```

```python
datacustomer.head()
```

```python
datacustomer.shape
```

```python
datacustomer.isnull().sum()
```

```python
import warnings
warnings.filterwarnings('ignore')

plt.figure(1 , figsize = (15 , 6))
n = 0 
for x in ['Age' , 'Annual Income (k$)' , 'Spending Score (1-100)']:
    n += 1
    plt.subplot(1 , 3 , n)
    plt.subplots_adjust(hspace =0.5 , wspace = 0.5)
    sns.distplot(datacustomer[x] , bins = 20)
    plt.title('Distplot of {}'.format(x))
plt.show()
```

```python
labels = ['Female', 'Male']
size = datacustomer['Genre'].value_counts()
colors = ['green', 'orange']
explode = [0, 0.1]

plt.rcParams['figure.figsize'] = (9, 9)
plt.pie(size, colors = colors, explode = explode, labels = labels, shadow = True, autopct = '%.2f%%')
plt.title('Gender', fontsize = 20)
plt.axis('off')
plt.legend()
plt.show()
```

```python
plt.figure(1 , figsize = (15 , 7))
n = 0 
for cols in ['Age' , 'Annual Income (k$)' , 'Spending Score (1-100)']:
    n += 1 
    plt.subplot(1 , 3 , n)
    plt.subplots_adjust(hspace = 0.5 , wspace = 0.5)
    sns.violinplot(x = cols , y = 'Genre' , data = datacustomer , palette = 'vlag')
    sns.swarmplot(x = cols , y = 'Genre' , data = datacustomer)
    plt.ylabel('Genre' if n == 1 else '')
    plt.title('Violin and Swarmplots' if n == 2 else '')
plt.show()
```

```python
#Clustering Based on Annual Income and Spending Score
X = datacustomer.iloc[:,[2,3]].values
X.shape
```

```python
wcss = []

for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, random_state=10)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)

print(wcss)
```

```python
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Clusters')
plt.ylabel('Inertia')
```

Nilai k optimal = 5

```python
kmeans = KMeans(n_clusters = 5, random_state = 10)
model = kmeans.fit(X)
```

```python
model.labels_
```

```python
model.cluster_centers_
```

```python
y_kmeans = kmeans.predict(X)
```

```python
#Cluster Scatter Plot
plt.scatter(X[y_kmeans == 0, 0], X[y_kmeans == 0, 1], s=70, c='red', label ='Cluster 1')
plt.scatter(X[y_kmeans == 1, 0], X[y_kmeans == 1, 1], s=70, c='blue', label ='Cluster 2')
plt.scatter(X[y_kmeans == 2, 0], X[y_kmeans == 2, 1], s=70, c='green', label ='Cluster 3')
plt.scatter(X[y_kmeans == 3, 0], X[y_kmeans == 3, 1], s=70, c='cyan', label ='Cluster 4')
plt.scatter(X[y_kmeans == 4, 0], X[y_kmeans == 4, 1], s=70, c='magenta', label ='Cluster 5')
plt.scatter(model.cluster_centers_[:, 0], model.cluster_centers_[:, 1], s=50, c='yellow', label = 'Centroids')
plt.title('Cluster of Customers Based on Annual Income and Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
```
